Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x) = x^2 + 5$ 

1. x > -1 oraz x < 1
2. x > -6 oraz x < 6
3. x > 0 oraz x < 5

Wspierając się dokumentacją Matplotlib. Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

Wskazówki: Deklaracja funkcji w Python.

W Pythonie bloki kodu (w tym również funkcje) są wyróżniane za pomocą wcięć. Do wizualizacji wyników w Pythonie używa się pakietu Matplotlib.

W pliku mają być zawarte: imię, nazwisko, opis wykonywanego ćwiczenia - formatowanie markdown w poszczególnych komórkach.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def f1 (x):
    return (x*x +5)

x1 = np.linspace(-1,1,100)
y1 = f1(x1) 

x2 = np.linspace(-6,6,100)
y2 = f1(x2)

x3 = np.linspace(0,5,100)
y3 = f1(x3)

plt.plot(x1, y1, "c", label = "Wykres w przedziale <-1,1>")
plt.grid()
plt.title("Wykres x^2 + 5 w przedziale <-1,1>")
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.legend()
plt.show()

plt.plot(x2, y2, "m", label = "Wykres w przedziale <-6,6>")
plt.grid()
plt.title("Wykres x^2 + 5 w przedziale <-6,6>")
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.legend()
plt.show()

plt.plot(x3, y3, "r", label = "Wykres w przedziale <0,5>")
plt.grid()
plt.title("Wykres x^2 + 5 w przedziale <0,5>")
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.legend()
plt.show()
```

**Wykonanie ćwiczenia**

Ćwiczenie zostało wykonane dzięki funkcji linspace z biblioteki numpy, dzięki której został stworzony  100-elementowy "wektor" liczb w danym przedziale. Następnie każdy składnik został argumentem funkcji $f(x) = x^2 + 5$. Wyniki tej operacji zostały wyświetlone na odpowiedich wykresach. Każdy wykres został wyświetlony w innym oknie ze względu na czytelność odczytu otrzymanych wyników.

Utwórz dataframe, w którym kolumny mają nazwy: name, surname, age, sex. Uzupełnij pięcioma dowolnymi rekordami oraz wyświetl informacje o danych pandas .info(), opis danych pandas .describe(), wyświetl pierwsze trzy rekordy pandas .head().

```python
import pandas as pd
data1 = {'Name': ["Adam", "Andrzej", "Aleksandra", "Amelia","Anastazy"], 
         'Surname': ["Iwaszkiewicz","Ibowicz","Iciachowska","Idziak","Ignasiak"], 
         "Age": [16,71,5,26,31], "Sex":["male","male","female","female","male"]}

dataframe1 = pd.DataFrame(data=data1, index=[1, 2, 3, 4,5])
print(dataframe1)
```

```python
dataframe1.info()
```

```python
dataframe1.describe(include="all") 
```

```python
dataframe1.head(n=3)
```

Metoda $info$ drukuje informacje o dataframe takie jak typ, kolumny, wartości inne niż null oraz użycie pamięci.

Metoda $describe$ generuje statystyki opisowe podsumowujące centralną tendencję, rozproszenie i rozkład zbioru danych, z wyłączeniem wartości NaN. Analizuje zarówno serie liczbowe, jak i obiektowe. Dane wyjściowe będą się różnić w zależności od tego, jakie są dostarczone dane wejściowe.
